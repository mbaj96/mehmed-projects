**Introduction**

In this project, we will be exploring a dataset on the job outcomes of students who graduated from college between 2010 and 2012.  We will aim to use data visualisation to answer a number of questions from the dataset, such as:

* Do students in more popular majors make more money? (using scatter plots)


* How many majors are predominantly male, and how many are predominantly female? (using histograms)


* Which category of majors have the most students? (using bar plots)

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.iloc[0]

```

**Generating Scatter Plots**

```python
plt.scatter(recent_grads['Sample_size'], recent_grads['Median'])

plt.xlabel('Sample Size')
plt.ylabel('Median Income')
plt.show()
```

```python
plt.scatter(recent_grads['Sample_size'], recent_grads['Unemployment_rate'])

plt.xlabel('Sample Size')
plt.ylabel('Median Income')
plt.show()
```

```python
plt.scatter(recent_grads['Full_time'], recent_grads['Median'])

plt.xlabel('Full Time')
plt.ylabel('Median Income')
plt.show()
```

```python
plt.scatter(recent_grads['ShareWomen'], recent_grads['Unemployment_rate'])

plt.xlabel('Share of Women')
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
plt.scatter(recent_grads['Men'], recent_grads['Median'])

plt.xlabel('Men')
plt.ylabel('Median Income')
plt.show()
```

```python
plt.scatter(recent_grads['Women'], recent_grads['Median'])

plt.xlabel('Women')
plt.ylabel('Median Income')
plt.show()
```

Using these plots we will answer the following questions:

* **Do students in more popular majors make more money?**


We can see that doesn't seem to be the case.  It seems as though the majors with a smaller sample size are the ones in which graduates have a higher median income.

* **Do students that majored in subjects that were majority female make more money?**


It seems as though female graduates tend to have a higher salary than male graduates, but I don't believe these graphs are enough to answer this question.


* **Is there any link between the number of full-time employees and median salary?**


There does seem to be - the smaller the number of full time employees, the higher the  median salary.



**Generating Histograms**

```python
cols = ['Sample_size', 'Median', 'Employed', 'Full_time', 
        'ShareWomen', 'Unemployment_rate', 'Men', 'Women']

fig = plt.figure(figsize=(10, 30))

for i in range(0,8):
    ax = fig.add_subplot(8,1,i+1)
    plt.xlabel(cols[i])
    ax = recent_grads[cols[i]].plot(kind='hist', bins=40)
    

```

Some observations here are:

* Most majors have less than 500 students who studied in them

* Median Salary lies somewhere between 30,000 and 40,000 dollars

* Unemployment rate for students is mostly distributed between 5-10%

**Creating Bar Plots**

```python
recent_grads[:10].plot.bar(x='Major', y ='ShareWomen', legend=False)
```

```python
recent_grads[163:].plot.bar(x='Major', y ='ShareWomen', legend=False)
```

From these plots we can determine that engineering subjects have a much lower percentage of female students than in fields such as astrophysics, psychology and education. 

```python
recent_grads[:10].plot.bar(x='Major', y ='Unemployment_rate', legend=False)
```

```python
recent_grads[163:].plot.bar(x='Major', y ='Unemployment_rate', legend=False)
```

We can see when comparing the unemployment rates for each major that there isn't a clear relationship between the type of major and the unemployment rate, at least for the majors displayed. We can see that Clinical Psychology and Nuclear Engineering have higher than average unemployment rates.  Other types of engineering such as Petroleum Engineering on the other hand have lower unemployment rates, alongside Astrophysics and Early Childhood Education.
